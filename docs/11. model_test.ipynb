````
import os

from dotenv import load_dotenv
from openai import OpenAI

# .env 파일에 OPENAI_API_KEY=sk-xxxxxx 저장되어 있어야 합니다.
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

MODEL_ID = "ft:gpt-4.1-mini-2025-04-14:skn14:hjk-0911-v01:CEUOn4Sy"

# 간단한 질의 함수
def ask_model(question: str) -> str:
    response = client.chat.completions.create(
        model=MODEL_ID,
        messages=[
            {"role": "system", "content":
                "너는 홍진경의 말투와 생각을 그대로 따라하는 챗봇이야.  패션에 관심이 많은 모델이며 아낌없는 조언을 하지. "
                "과장된 인터넷체 금지. 핵심을 먼저"
                "적절히 반문과 리액션 섞기. 사실관계가 필요하면 솔직하게 모른다고 말하고, 확인 질문하기."},
            {"role": "user", "content": question}
        ],
        temperature=0.9,
        top_p=0.8,
    )
    return response.choices[0].message.content

````

````
# 일반적인 질의(말투 확인)
ask_model("안녕? 오늘 날씨에 맞는 코디 추천해줘.")
````
